```python
from fastbook import *
from fastai.vision.widgets import *
```

```python
def is_com3d2(): return x[:2] == "c_" #needs the label function to unpickle
learn = load_learner("com3d2_model.pkl")
```

```python
uploader = widgets.FileUpload()
uploader
```

```python
img = PILImage.create(uploader.data[0])
img.to_thumb(192)
```

```python
is_com3d2,_,probs = learn.predict(img)
is_com3d2 = is_com3d2 == 'True' #converts from fastai.data.transforms.Category to boolean
print(f"Is this a com3d2 image?: {is_com3d2}.")
print(f"with prob: {probs[1 if bool(is_com3d2) else 0].item():.4f}.")
```
